```python
file_path = '/content/drive/My Drive/DataProjects/Employee.csv'  # Adjust if the filename differs
df = pd.read_csv(file_path)
df.head()
```

```python
print(df.isnull().sum())

# Drop rows with missing values (if any)
df = df.dropna()

# Verify Data Types
print(df.dtypes)
```

```python
print(df.columns)
```
